Terminology
- A model implements a machine learning algorithm
- In scikit-learn, models are called estimators. 
- Estimaters in scikit learn are classes that implements ML model based on specific algorithms
- Two parameter types in machine learning
    - Those the estimator calculates during the ML process.
    - Those you specify in advance when you create the estimator object.
        - so these parameters specified in advance are called hyperparameters
        - In k-nearest neighbors, k is a hyperparameter
        - for simplicity we will be using default parameter values
        - In real world machine-learning studies, you will want to experiment with different values of k
- scikit-learn also has automated hyperparameter tuning capacbilities.
