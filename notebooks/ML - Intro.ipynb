Supervised Learning
    - Data is already labeled.
    - Someone has already told us what that data is
    - Our algorithm is going to learn from that data using a subset of it called training
    - Traing data has access to data samples and targets(values that this data represents)
    - Model trains itself on this labeled data and learns about this data
    - Then we provide data which model has not seen before and then Model use what it has learned to figure out the value (or label ) for that data.
    - We can check the accuracy of that prediction by matching it with labels we already have from the dataset.
    - Once required level of acuracy is achieved then we can use this model on data never seen or labeled before.

    - You have unlebeled data and you do not know what that data looks like
