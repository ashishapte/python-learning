```python
from sklearn.datasets import fetch_california_housing
```

```python
california = fetch_california_housing()
# you get a bunch object with california housing dataset
```

```python
california.data.shape
```

```python
california.target.shape
```

```python
california.feature_names
```

```python
import pandas as pd
```

```python
pd.set_option('precision', 4)
```

```python
pd.set_option('max_columns', 9)
```

```python
pd.set_option('display.width', None)
```

```python
california_df = pd.DataFrame(california.data, columns=california.feature_names)
```

```python
california_df['MedHouseValue'] = pd.Series(california.target)
```

```python
california_df.head()
```

```python
california_df.describe()
```

```python
california_df.shape
```

```python
# we can create fraction of this dataset using pandas 
# use that fractional dataset for visualization
sample_df = california_df.sample(frac=0.1, random_state=17)
```

```python
sample_df.shape
```

```python
import matplotlib.pyplot as plt
# used to create figure object and dispaly on the screen
```

```python
import seaborn as sns
# use seaborn to create a scatterplot
```

```python
sns.set_style('whitegrid')
# set whitegrid background for better display
```

```python
plt.figure(figsize=(16,9))
```

```python
for feature in california.feature_names:
    plt.figure(figsize=(16,9)) # 16-by-9 inch figure
    sns.scatterplot(data=sample_df, x=feature, y='MedHouseValue', palette='cool', legend=False)
```
