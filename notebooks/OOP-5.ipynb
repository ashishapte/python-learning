```python
#Adding everything in one cell for Jupyter to work
# ravel allows you to run through multi dimentional array as if you are delaing with single dimentational array.
# you may run into None object type or similar error - this is because counter is getting increased as we deal cards.
# its not resetting .. so after 52 there are no more cards in _cards list/
figure, axes_list = plt.subplots(nrows=4, ncols=13)
figure.set_figwidth(16)
figure.set_figheight(9)
for axes in axes_list.ravel():
    axes.get_xaxis().set_visible(False)
    axes.get_yaxis().set_visible(False)
    image_name = deck_of_cards.deal_card().image_name.lower()
    img = mpimg.imread(str(path.joinpath(image_name).resolve()))
    axes.imshow(img)
    
figure.tight_layout() # uses available space optimally
```
