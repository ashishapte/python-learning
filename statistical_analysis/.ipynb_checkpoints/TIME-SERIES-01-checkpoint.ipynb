Time-Series are datasets where observations are arranged in chronological order




If T is set to real numbers, its continuous-time stochastic process
If T is set to integers, we call it stochastic process in discrete time.

Characterstics 
    - Multivariate time series
    - Univariate series
    
Imp Char:
- Long term movements of the values (trend)
- Seasonal variation (seasonality) - variation that occurs at specific interval of less than a year.
    e.g. sale peaking on each weekend. This could be daily, weekly, monthly or yearly
- Irregular or cyclic components  - This is variability thats not of a fixed frequency or that rises and falls in a way thats not based on seaonal frequency.
- Noise


Stationarity
- is the property of a time-series not to change its distribution over time as described by its summary statistics.
- If the time-series is stationary, it means that it has no tread and no deterministic seasonal variability , although other cyclical variability is permitted. This is important for algorithms - Forecasting with Moving averages and autoregressive models. To apply that, we need to transform non stationary data into stationary data by removing seasonality and trend

The task of identifying, quantifying, and decomposing these and other characteristics is calledÂ time-series analysis. Exploratory time-series analysis is often the first step before any feature transformation and machine learning
