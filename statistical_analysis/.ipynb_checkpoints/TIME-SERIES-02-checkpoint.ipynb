Time Series Analysis

Definition: Time-Series Analysis (TSA) is the process of extracting a summary and other statistical information from time-series, most importantly, the analysis of trend and seasonality


Descriptive analysis : summarizes characteristics of a dataset
Exploratory analysis : analyzes for patterns, trends, or relationships between variables


Steps in working with Time Series data
- Importing the dataset
- Data Cleaning 
- Understanding variables
- Uncovering relationships between variables
- Indentifying trend and seasonality
- Preprocessing (including feature engineering)
- Training a machine learning model

Above steps are highly iterative



```python
Data Anslysis - 
- what is the size of the dataset (number of observations) and its shape
- How many features or columns we have
- what are the column types
- Look at histograms and distribution plots - for assessing relationship between features and target variables
- calculate correlations and visualize as a correlation heatmap.
- Look for missing values and clean up and correct these irregularities
- Investigate relationship of time with each of the variable
- analyze relationships between variables

Univariate analysis : 
    - Look at single variable. 
    - Inspect values to get means and variance.
    - or graphically plot the distribution

Multivariate analysis:
    - calculating correaltions between variables
    - graphically drawing scatter plot 
    
```

The standard error (SE) is an approximation of the standard deviation of sampled data. 
It measures the dispersion of sample means around the population mean, 
but normalized by the root of the sample size. 
The more data points involved in the calculation, the smaller the standard error tends to be.
